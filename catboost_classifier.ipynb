```python
import pandas as pd
import os
import numpy as np
np.set_printoptions(precision=4)
import catboost
from catboost import *
from catboost import datasets, Pool, CatBoostClassifier
```

```python
!wget https://raw.githubusercontent.com/vsecoder/ru-toxic-messages-classification/main/labeled.csv
df = pd.read_csv('labeled.csv')
```

```python
df.head()
```

```python
# оставим 1000 на тесты, остальным обучим
x = len(df) - 1000
train_df, test_df = df[:x], df[x:]
```

```python
train_df = {
    'comment': train_df.comment,
    'toxic': [int(i) for i in train_df.toxic]
}

train_df = pd.DataFrame(train_df)

train_df_ = train_df['toxic']
train_df = train_df.drop(['toxic'], axis=1)
```

```python
test_df = {
    'comment': test_df.comment,
    'toxic': [int(i) for i in test_df.toxic]
}

test_df = pd.DataFrame(test_df)

test_df_ = test_df['toxic']
test_df = test_df.drop(['toxic'], axis=1)
```

```python
pool = Pool(train_df, train_df_, text_features=['comment'])
test_pool = Pool(test_df, test_df_, text_features=['comment'])
```

```python
model = CatBoostClassifier(
    iterations=800,
    learning_rate=0.1
    eval_metric='Accuracy'
)

model.fit(pool, verbose=50, eval_set=test_pool)

print('Model is fitted: ' + str(model.is_fitted()))
print('Model params:')
print(model.get_params())
```

```python
print(model.predict(pd.DataFrame({'comment': ['Дуй спать, малолетка']})))
```
